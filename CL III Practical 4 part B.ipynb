## Problem Statement : Implement Ant colony optimization by solving the Traveling salesman problem using python

```python
import numpy as np
import random
```

```python
# Distance Matrix (TSP)
# -----------------------------------
distance_matrix = np.array([
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
])

num_cities = len(distance_matrix)
```

```python
# Parameters
# -----------------------------------
num_ants = 10
num_iterations = 50
alpha = 1        # pheromone importance
beta = 2         # visibility importance
evaporation_rate = 0.5
```

```python
# Initialize pheromone & visibility
# -----------------------------------
pheromone = np.ones((num_cities, num_cities))

# Avoid divide by zero
visibility = 1 / (distance_matrix + np.eye(num_cities))
```

```python
# Function to calculate route distance
# -----------------------------------
def calculate_distance(route):
    distance = 0
    for i in range(len(route) - 1):
        distance += distance_matrix[route[i]][route[i+1]]
    # return to start
    distance += distance_matrix[route[-1]][route[0]]
    return distance
```

```python
# ACO Algorithm
# -----------------------------------
best_route = None
best_distance = float('inf')

for iteration in range(num_iterations):
    all_routes = []
    all_distances = []

    for ant in range(num_ants):
        current_city = random.randint(0, num_cities - 1)
        visited = [current_city]

        while len(visited) < num_cities:
            probabilities = []
            total = 0

            for city in range(num_cities):
                if city not in visited:
                    tau = pheromone[current_city][city] ** alpha
                    eta = visibility[current_city][city] ** beta
                    prob = tau * eta
                    probabilities.append((city, prob))
                    total += prob

            # Normalize probabilities
            probabilities = [(city, prob / total) for city, prob in probabilities]

            # Roulette wheel selection
            r = random.random()
            cumulative = 0
            for city, prob in probabilities:
                cumulative += prob
                if r <= cumulative:
                    next_city = city
                    break

            visited.append(next_city)
            current_city = next_city

        route = visited
        dist = calculate_distance(route)

        all_routes.append(route)
        all_distances.append(dist)

        # Update best solution
        if dist < best_distance:
            best_distance = dist
            best_route = route
    # Pheromone Update
    # -----------------------------------
    pheromone = (1 - evaporation_rate) * pheromone

    for route, dist in zip(all_routes, all_distances):
        for i in range(len(route) - 1):
            a = route[i]
            b = route[i+1]
            pheromone[a][b] += 1 / dist
            pheromone[b][a] += 1 / dist

    # Print progress
    print(f"Iteration {iteration+1}: Best Distance = {best_distance}")
```

```python
# Final Result
# -----------------------------------
print("\nBest Route:", best_route)
print("Shortest Distance:", best_distance)
```
